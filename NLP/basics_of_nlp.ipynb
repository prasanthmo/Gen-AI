```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.npr.org/2021/02/10/966396848/read-trumps-jan-6-speech-a-key-part-of-impeachment-trial'
response = requests.get(url)

# Parse the HTML content of the response
soup = BeautifulSoup(response.content, 'html.parser')
print(soup)
paragraphs = soup.find_all('p')
transcript = ''
for paragraph in paragraphs:
    transcript += paragraph.get_text() + '\n'

# The transcript is now stored in the 'transcript' variable
# You can use it later in your code, for example:
print(transcript)  # Print the transcript
# ... other operations using the 'transcript' variable ...
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('punkt')
```

```python

import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# Assuming 'transcript' is available from the previous cell
tokens = word_tokenize(transcript)

print(tokens)
```

```python
tokens = word_tokenize(transcript)
tokens = [token.lower() for token in tokens]  # Lowercase the tokens

print(tokens)
```

```python

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

# Assuming 'tokens' is available from the previous cell
filtered_tokens = [w for w in tokens if not w.lower() in stop_words]

print(filtered_tokens)
```

```python

import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

 # Replace with your actual filtered tokens filtered_tokens

# Initialize Porter Stemmer
ps = PorterStemmer()

# Stem the tokens
stemmed_tokens = [ps.stem(token) for token in filtered_tokens]

print(stemmed_tokens)
# Print the number of tokens
print("Number of tokens:", len(stemmed_tokens))
```

```python

import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer


# Assuming 'tokens' is available from the previous cell
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

print(lemmatized_tokens)
# Print the number of tokens
print("Number of tokens:", len(lemmatized_tokens))
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
sentiment_scores = sia.polarity_scores(transcript)
print(sentiment_scores)
if sentiment_scores['compound'] >= 0.05:
    sentiment = 'Positive'
elif sentiment_scores['compound'] <= -0.05:
    sentiment = 'Negative'
else:
    sentiment = 'Neutral'
print("Sentiment:", sentiment)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a list of documents (in this case, just one document)
documents = [' '.join(lemmatized_tokens)]  # Ensure `lemmatized_tokens` is defined

# Create a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# Fit the vectorizer to the documents and transform them into a TF-IDF representation
tfidf_matrix = vectorizer.fit_transform(documents)

# Get feature names and TF-IDF values
feature_names = vectorizer.get_feature_names_out()
tfidf_values = tfidf_matrix.toarray()[0]  # Get values for the first document

# Sort by TF-IDF values in descending order
sorted_indices = np.argsort(tfidf_values)[::-1]  # Get indices of sorted values
sorted_feature_names = [feature_names[i] for i in sorted_indices]
sorted_tfidf_values = [tfidf_values[i] for i in sorted_indices]

# Select top N words to display (e.g., top 10)
top_n = 10
top_feature_names = sorted_feature_names[:top_n]
top_tfidf_values = sorted_tfidf_values[:top_n]

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_feature_names, top_tfidf_values, color='skyblue')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.xlabel("Words")
plt.ylabel("TF-IDF Score")
plt.title("Top 10 Words by TF-IDF Score")
plt.show()
```
