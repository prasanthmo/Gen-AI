```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Assuming `my_track_sets` is the binary matrix where rows are customer_id and columns are track_id
# Each value is 1 if listened to, 0 if not listened

# Step 1: Compute User Similarity Matrix using Cosine Similarity
user_similarity = pd.DataFrame(
    cosine_similarity(my_track_sets),
    index=my_track_sets.index,
    columns=my_track_sets.index
)

# Step 2: Define Collaborative Filtering Recommendation Function
def collaborative_recommendations(customer_id, my_track_sets, user_similarity, top_n=5):
    """
    Recommend tracks based on collaborative filtering using cosine similarity.

    Parameters:
    - customer_id: ID of the customer to make recommendations for.
    - my_track_sets: Binary matrix of user-track interactions.
    - user_similarity: Precomputed user similarity matrix.
    - top_n: Number of recommendations to return.

    Returns:
    - List of recommended track IDs or a message if customer_id is not found.
    """
    # Check if the customer_id exists in the dataset
    if customer_id not in my_track_sets.index:
        print(f"Customer ID {customer_id} is not found in the dataset.")
        return []

    # Find the most similar users to the given customer_id
    similar_users = user_similarity[customer_id].sort_values(ascending=False).drop(customer_id).head(top_n)

    # Tracks the target customer has already listened to
    user_listened_tracks = set(my_track_sets.loc[customer_id][my_track_sets.loc[customer_id] > 0].index)

    # Collect tracks from similar users that the target customer hasn't listened to
    recommended_tracks = {}
    for similar_user in similar_users.index:
        # Get tracks listened to by similar users
        similar_user_tracks = my_track_sets.loc[similar_user]

        # Recommend tracks the target customer has not yet listened to
        for track in similar_user_tracks[similar_user_tracks > 0].index:
            if track not in user_listened_tracks:  # Recommend only if the user hasn't listened
                # Weight each recommendation by the similarity score of the similar user
                if track in recommended_tracks:
                    recommended_tracks[track] += similar_users[similar_user]
                else:
                    recommended_tracks[track] = similar_users[similar_user]

    # Sort tracks by summed similarity score and get top N recommendations
    recommended_tracks = sorted(recommended_tracks.items(), key=lambda x: x[1], reverse=True)
    recommended_tracks = [track for track, score in recommended_tracks][:top_n]

    return recommended_tracks

# Prompt user for customer_id input
try:
    customer_id = int(input("Enter Customer ID for recommendations: "))
    # Generate recommendations for the entered customer_id
    recommendations = collaborative_recommendations(customer_id, my_track_sets, user_similarity, top_n=5)

    # Display the recommendations or message if no recommendations are available
    if recommendations:
        print(f"Recommended Tracks for Customer {customer_id}:", recommendations)
    else:
        print(f"No recommendations available for Customer {customer_id}.")
except ValueError:
    print("Invalid input. Please enter a numeric Customer ID.")
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.metrics.pairwise import cosine_similarity

# Assuming `my_track_sets` is the binary matrix where rows are customer_id and columns are track_id
# Convert `my_track_sets` to boolean to avoid deprecation warnings for apriori
my_track_sets = my_track_sets.astype(bool)

# Step 1: Run the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(my_track_sets, min_support=0.01, use_colnames=True)

# Generate association rules from the frequent itemsets
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Step 2: Compute User Similarity Matrix for Collaborative Filtering
user_similarity = pd.DataFrame(
    cosine_similarity(my_track_sets),
    index=my_track_sets.index,
    columns=my_track_sets.index
)

# Define a function for content-based recommendations (association rules)
def get_content_based_recommendations(customer_id, my_track_sets, rules, top_n=5):
    if customer_id not in my_track_sets.index:
        print(f"Customer ID {customer_id} is not found in the dataset.")
        return []

    user_tracks = set(my_track_sets.loc[customer_id][my_track_sets.loc[customer_id] > 0].index)
    recommendations = rules[rules['antecedents'].apply(lambda x: x.issubset(user_tracks))]
    recommendations = recommendations.sort_values(['confidence', 'lift'], ascending=[False, False])['consequents'].head(top_n)
    recommended_tracks = [item for sublist in recommendations for item in sublist if item not in user_tracks][:top_n]

    return recommended_tracks

# Define a function for collaborative filtering recommendations (cosine similarity)
def get_collaborative_recommendations(customer_id, my_track_sets, user_similarity, top_n=5):
    if customer_id not in my_track_sets.index:
        print(f"Customer ID {customer_id} is not found in the dataset.")
        return []

    similar_users = user_similarity[customer_id].sort_values(ascending=False).drop(customer_id).head(top_n)
    user_listened_tracks = set(my_track_sets.loc[customer_id][my_track_sets.loc[customer_id] > 0].index)

    recommended_tracks = {}
    for similar_user in similar_users.index:
        similar_user_tracks = my_track_sets.loc[similar_user]
        for track in similar_user_tracks[similar_user_tracks > 0].index:
            if track not in user_listened_tracks:
                if track in recommended_tracks:
                    recommended_tracks[track] += similar_users[similar_user]
                else:
                    recommended_tracks[track] = similar_users[similar_user]

    recommended_tracks = sorted(recommended_tracks.items(), key=lambda x: x[1], reverse=True)
    recommended_tracks = [track for track, score in recommended_tracks][:top_n]

    return recommended_tracks

# Prompt user for customer_id input and generate combined recommendations
try:
    customer_id = int(input("Enter Customer ID for recommendations: "))

    # Generate content-based recommendations
    content_recommendations = get_content_based_recommendations(customer_id, my_track_sets, rules, top_n=5)
    # Generate collaborative filtering recommendations
    collaborative_recommendations = get_collaborative_recommendations(customer_id, my_track_sets, user_similarity, top_n=5)

    # Combine recommendations and remove duplicates
    combined_recommendations = list(set(content_recommendations + collaborative_recommendations))

    # Display the combined recommendations
    if combined_recommendations:
        print(f"Combined Recommendations for Customer {customer_id}:", combined_recommendations)
    else:
        print(f"No recommendations available for Customer {customer_id}.")

except ValueError:
    print("Invalid input. Please enter a numeric Customer ID.")
```
